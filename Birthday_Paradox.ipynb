## Birthday Paradox

Birthday paradox concerns the probability that, in a set of n randomly chosen people, some pair of them will have the same birthday. You can read about this problem at https://en.wikipedia.org/wiki/Birthday_problem.

```python
def get_duplicate_list(students):
    import random
    
    #initialize 365-day year list
    year = [0] * 365
    
    #initialize duplicate list
    duplicate_list = list()  
    
    #Iterate each student...
    for i in range(students):
        #get random birthday
        birthday =  random.randrange(365)
        
        #Assign random BD to year array index.
        year[birthday] = year[birthday] + 1

        #if duplicate found, append to the list
        if year[birthday] > 1:
            duplicate_list.append(birthday)
    
    #return duplicate list
    return duplicate_list

def calc_probability(samples, students):   
    
    #initialize dictionary
    duplicate_dict = dict()
    
    #Iterate each sample
    for sample in range(samples):
     
        #get duplicate list
        duplicate_list = get_duplicate_list(students)
        
        #if any duplicate found...
        if len(duplicate_list) > 0:
            #add duplicate list to dictionary along with sample as key
            duplicate_dict.update({sample: duplicate_list})
    
    print('number of students', students)
    print('number of iterations/samples', samples)
    
    #Iterate and print each duplicate and print duplicate
    print('list of duplicate days for each iteration, where duplicates are found')
    for i in duplicate_dict:
        print('\tIteration #', i + 1, duplicate_dict[i])
        
    #calculate probability for two or more students is that total duplicates divide by number iteration/samples
    print('probability of two students shared the same birthday is', float(len(duplicate_dict) / samples))
    
    #calculate probabilities of non-shared/shared birthdays (**little extra from assignment**)
    probablity_non_shared = 1.0
    for i in range(students):
        probablity_non_shared = probablity_non_shared * (365 - i) / 365

    print("overall probability that there are no shared birthdays:", round(probablity_non_shared, 2))
    print("overall probability that there are shared birthdays:", round((1 - probablity_non_shared), 2))

#provide sample/iteration in first argument and then number of students in second
calc_probability(1000, 45)
```
