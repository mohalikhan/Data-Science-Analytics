```python
# Exercise #1: Create a plot where y = x**2 for x over the interval 1 to 5, properly labelled
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
%matplotlib inline

# Start here

# Create x list from 1 to 5
x = range(1, 6)

# get power by 2 for each value in x
y = [x*x for x in x]

# set chart lables
plt.title("Excercise # 1")
plt.xlabel("X Value")
plt.ylabel("Y Value")

# create line chart
plt.plot(x, y, '-ok', color='green', alpha=0.6)
```

```python
# Exercise #2: Create a histogram where the mean = 0, std. dev. = 1, n = 300, and there are sqrt(n) bins
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
%matplotlib inline 

# Start here
mean, stdDev, n = 0, 1, 300 # mean and stadard deviation and number of count

# get data by random number
x = np.random.normal(mean, stdDev, n)

# get number of bins by sqrt(n) as integer
num_bins = np.sqrt(n).astype(int);

# the histogram of the data
count, bins, patches = plt.hist(x, num_bins, normed=True, label='Std. Deviation')

# add a 'best fit' line
y = mlab.normpdf(bins, mean, stdDev)

# add plot with custom line
plt.plot(bins, y, '--', alpha=.95, color='black')

# set titles
plt.title('Excercise # 2')
plt.xlabel('Outcome')
plt.ylabel('Frequency')

# layout and show
plt.tight_layout()
plt.legend()
plt.show()

# The chart will be a little different each time you run it
```

```python
# Exercise #3: (1) Create a line plot of your choosing with an appropriate legend which displays
# the formula of the curve depicted. (2) The formula should be formatted using TeX syntax
# to create a professional presentation. (3) Finally save your plot as a PDF file. 
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 

# get the spaced numbers
x = np.linspace(-1, 4, 50)

# apply y to sine function of x
y = np.sin(x)

# set chart labels and title
plt.title(r"Excercise # 3",
          fontsize=12, color='black', fontweight='bold')
plt.xlabel("X Value")
plt.ylabel("Y Value")

# create chart and apply TeX syntax to label
plt.plot(x, y, color='red', alpha=0.75, label="$y = \sin \\theta$")
plt.legend()

# save to pdf
plt.savefig("Exercise3.pdf") 

# show chart
plt.show()
```
